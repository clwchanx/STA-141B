```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
df = pd.read_csv('finaldata.csv')
df = df.iloc[:,[1,2,3,4,5,6,8]]
df
```

```python
df['Benefits'] = df['Benefits'].replace({'Yes':1,'No':0})
df['Machine Learning'] = df['Machine Learning'].replace({'Yes':1,'No':0})
df['Type'] = df['Type'].replace({'Scientist':1,'Analyst':0})
df['Education'] = df['Education'].replace({'Unspecified':0,'Bachelor':1,'Master':2,'PhD':3})
```

```python
normal = preprocessing.normalize(df)
```

```python
kmeans = KMeans(n_clusters = 2, random_state = 0)
clusters = kmeans.fit(normal)
```

```python
df['Cluster'] = clusters.labels_
```

```python
df['Benefits'] = df['Benefits'].replace({1:'Yes',0:'No'})
df['Machine Learning'] = df['Machine Learning'].replace({1:'Yes',0:'No'})
df['Type'] = df['Type'].replace({1:'Scientist',})
df['Education'] = df['Education'].replace({'Unspecified':0,'Bachelor':1,'Master':2,'PhD':3})
```
